# Merging dataframes

```python
# dropping existing _merge column
df_ords_prods.drop(columns = ['_merge'], inplace = True)
```

```python
# merging df_ords_prods and df_prods dataframe using product_id as key, add a merge flag
df_merged = df_ords_prods.merge(df_prods, on = 'product_id', indicator = True)
```

```python
# confirming the merge using the _merge flag
df_merged['_merge'].value_counts()
```

# Export the merged dataframe
